```python
import pandas as pd
import matplotlib.pyplot as plt

weather_df = pd.read_csv('/Users/crobinson1205/WebDesign_HW/assets/weather_cities.csv')
weather_df.head()
weather_df.to_html('WeatherData.html')
```

```python
temp = weather_df.plot.scatter('Lat','Max Temp',color = 'Navy')
temp.set_xlim(-85,85)
temp.set_xlabel('Latitude')
temp.set_ylabel('Max Temperature(F)')
temp.set_title('City Latitude v Max Temperature')
plt.savefig('Lat_vs_Temp.jpeg')
```

```python
hum = weather_df.plot.scatter('Lat','Humidity',color = 'Navy')
hum.set_xlim(-85,85)
hum.set_xlabel('Latitude')
hum.set_ylabel('Humidity(%)')
hum.set_title('City Latitude v Humidity')
plt.savefig('Lat_vs_Hum.jpeg')
```

```python
cloud = weather_df.plot.scatter('Lat','Cloudiness',color = 'Navy')
cloud.set_xlim(-85,85)
cloud.set_xlabel('Latitude')
cloud.set_ylabel('Cloudiness(%)')
cloud.set_title('City Latitude v Cloudiness')
plt.savefig('Lat_vs_Cloudiness.jpeg')
```

```python
wind = weather_df.plot.scatter('Lat','Wind Speed',color = 'Navy')
wind.set_xlim(-85,85)
wind.set_xlabel('Latitude')
wind.set_ylabel('Wind Speed(mph)')
wind.set_title('City Latitude v Wind Speed')
plt.savefig('Lat_vs_Wind.jpeg')
```
